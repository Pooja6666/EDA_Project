```python
RAM= []
titels = []
Camera = []
Battery=[]
Storage= []
Prices= []
Offers = []
Ratings = []
Screen_size = []
Reviews = []

for i in range(1,100):
    
    url="https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    req.get(url,headers = {'User-Agent':'Mozilla/5.0'})
    headers = {"Accept-Language": "en-US,en;q=0.9"}
    page=req.get(url, headers=headers)
    
    soup = BeautifulSoup(page.text,"lxml")
    
    text_content_list = [element.get_text() for element in soup.find_all("div", class_="yKfJKb row")]
    for titels in soup.find_all('div',class_ = re.compile(r'KzDlHZ')):
        titels.append(titels.text.strip())
            
    for i in soup.find_all("div",class_="yKfJKb row"):
        if re.findall(r"Reviews(\d.)",i.text.strip()):
             RAM.append("".join(re.findall(r"Reviews(\d.)",i.text.strip())))
        else:
             RAM.append(np.nan)
        if re.findall(r"Display(\d+)",i.text.strip()):
            Camera.append(" ".join(re.findall(r"Display(\d+)",i.text.strip())))
        else:
            Camera.append(np.nan)
        if re.findall(r"Camera(\d+).mAh",i.text):
            Battery.append(" ".join(re.findall(r"Camera(\d+).mAh",i.text)))
        else:
            Battery.append(np.nan)
        if re.findall(r"(\d+).GB.ROM",i.text.strip()):
            Storage.append(" ".join(re.findall(r"(\d+).GB.ROM",i.text.strip())))
        else:
            Storage.append(np.nan)
        if re.findall(r'ccessories₹\d+.{4}₹(\d.{1,5})',i.text.strip()):
            Prices.append("".join(re.findall(r'ccessories₹\d+.{4}₹(\d.{1,5})',i.text.strip())))
        else:
            Prices.append(np.nan)
        if re.findall(r"(\d.)%",i.text):
            Offers.append("".join(re.findall(r"(\d.)%",i.text)))
        else:
            Offers.append(np.nan)
        if re.findall(r"GB.(\d.{2})",i.text.strip()):
            Ratings.append("".join(re.findall(r"GB.(\d.{2})",i.text.strip())))
        else:
            Ratings.append(np.nan)
        if re.findall(r"\W(\d.\d+)\s[a-z]",i.text.strip()):
            Screen_size.append("".join(re.findall(r"\W(\d.\d+)\s[a-z]",i.text.strip())))
        else:
            Screen_size.append(np.nan)
        if re.findall(r"&\s(\w.\d+).Reviews",i.text.strip()):
            Reviews.append("".join(re.findall(r"&\s(\w.\d+).Reviews",i.text.strip())))
        else:
            Reviews.append(np.nan)
            
df = pd.DataFrame({
    "RAM":RAM,
    "titels":titels,
    "Camera":Camera,
    "Battery":Battery,
    "Storage":Storage,
    "Prices":Prices,
    "Offers":Offers,
    "Ratings":Ratings,
    "Screen_size":Screen_size,
    "Reviews":Reviews    
})
```

```python
def format_and_split(model):
    # Remove extra spaces and commas
    model = re.sub(r'[ ,]+', ' ', model.strip())
    # Capitalize first letter of each word
    model = ' '.join(word.capitalize() for word in model.split())
    # Split into brand and model
    match = re.match(r'(^\w+) (.+)', model)
    if match:
        return pd.Series([match.group(1), match.group(2)])
    else:
        return pd.Series([model, ""])
#df = pd.DataFrame({'titles': titles})
df[['Brand', 'Model']] = df['titles'].apply(format_and_split)
df = df.drop(columns=['titles'])
df = df.drop_duplicates().sort_values(by=['Brand', 'Model']).reset_index(drop=True)
```

```python
df
```

```python

def clean_model(model):
    # Remove text inside brackets
    cleaned_model = re.sub(r'\s*\([^)]*\)', '', model).strip()
    return cleaned_model

# Apply the function to clean the 'Model' column
df['Cleaned_Model'] = df['Model'].apply(clean_model)


```
